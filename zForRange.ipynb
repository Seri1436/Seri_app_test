```python
import random as r    # random module

x=['Fruit','mango','rasberries','Blueberries','watermelon']    # the words in x
n=1   # to start n tells us what question number you're on
d=r.choice(x)   # d holds whatever word the computer chooses from x
input("enter to start")   # if you press the enter key then you start

while n<=5:
  print(n)    # it tells you the quesion number
  print(d)    # it tells you the word you have to spell
  g=input()   # put in the word

  if g==d:
    print("Good")   # if you get it right then you get to go to the next stage
    n=n+1   # your stage number adds 1 up
    d=r.choice(x)   # it chooses another word from x

  else:
    print("Try again")    # if you get it wrong, then you have to spell again

print("Great Work!")    # when it ends if says "the end"


```
